```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://jhr.uwpress.org/content/{}/{}.toc'
```

```python
result = []
```

```python
for i in [51,52]:
    for j in range(1,5):
        r = requests.get(url.format(i,j))
        soup = BeautifulSoup(r.content)
        metadata = soup.find_all("div", class_="cit-metadata")
        try:
            for tmp in metadata:
                j_meta = {}
                j_meta['doi'] = tmp.find("span", class_="cit-doi").text.replace("doi:", "")
                j_meta['Article_title'] = tmp.find("h4", class_="cit-title-group").text
                j_meta['Author(s)'] = tmp.find("ul", class_="cit-first-element cit-auth-list").text.replace("\n", "")
                j_meta['Publish date'] = tmp.find("span", class_="cit-print-date").text
                result.append(j_meta)
        except AttributeError:
            continue
```

```python
cols = ['doi', 'Article_title', 'Author(s)','Publish date']
df = pd.DataFrame(result)[cols]
df.head()
```
